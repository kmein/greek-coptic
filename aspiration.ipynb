```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option("styler.format.decimal", ",")
pd.set_option("styler.format.precision", 1)
pd.set_option("styler.format.na_rep", "–")

import matplotlib.pyplot as plt
cm = plt.cm.Blues
```

```python
df_kh = pd.read_csv("aspiration/kh.txt", sep="\t")
df_ph = pd.read_csv("aspiration/ph.txt", sep="\t")
df_th1 = pd.read_csv("aspiration/th1.txt", sep="\t")
df_th2 = pd.read_csv("aspiration/th2.txt", sep="\t")
df_th = pd.concat([df_th1, df_th2])
```

```python
df_th.loc[df_th["1_anno_scriptorium:norm"].isin(["ⲡⲉⲧ", "ⲣⲉϥⲣⲡⲉⲧ"]), "1_anno_scriptorium:norm"] = "ⲉⲧ"
```

```python
kh_full = pd.crosstab(df_kh["2_anno_scriptorium:norm"], df_kh["1_anno_scriptorium:norm"])
kh = df_kh[df_kh["1_span"].str[-1] == "ⲭ"]
kh_khi = pd.crosstab(kh["2_anno_scriptorium:norm"], kh["1_anno_scriptorium:norm"])
```

```python
ph = df_ph[df_ph["1_span"].str[-1] == "ⲫ"]
ph_phi = pd.crosstab(ph["2_anno_scriptorium:norm"], ph["1_anno_scriptorium:norm"])
ph_phi = ph_phi[ph_phi.sum(axis=1) > 1]
```

```python
ph_full = pd.crosstab(df_ph["2_anno_scriptorium:norm"], df_ph["1_anno_scriptorium:norm"])
ph_full = ph_full.loc[ph_phi.index, ph_phi.columns]
```

```python
th = df_th[df_th["1_span"].str[-1] == "ⲑ"]
th_theta = pd.crosstab(th["2_anno_scriptorium:norm"], th["1_anno_scriptorium:norm"])
th_theta = th_theta[th_theta.sum(axis=1) > 1]
```

```python
th_full = pd.crosstab(df_th["2_anno_scriptorium:norm"], df_th["1_anno_scriptorium:norm"])
th_full = th_full.loc[th_theta.index, th_theta.columns]
```

```python
(th_theta.div(th_full) * 100)
```

```python
th_theta.join(ph_phi, how="outer").join(kh_khi, how="outer").fillna(0).style.background_gradient(axis=None, cmap=cm)
```
